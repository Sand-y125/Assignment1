### Task 2 Composite Development Score:

##### i. Create a new metric called Composite Score using the formula:Composite Score = 0.30 Ã— Life Expectancy Index + 0.30 Ã— GNI per Capita Index

```python
south_asia_df = pd.read_csv('/content/drive/MyDrive/Assignment1/HDI_SouthAsia.csv')
south_asia_df
```

```python
south_asia_df[['life_expectancy', 'gross_inc_percap']].describe()   #to check indices or raw values
```

```python
south_asia_df['life_expectancy_index'] = ((south_asia_df['life_expectancy'] - south_asia_df['life_expectancy'].min())
                                          /(south_asia_df['life_expectancy'].max() - south_asia_df['life_expectancy'].min()))

south_asia_df['gni_index'] = ((south_asia_df['gross_inc_percap'] - south_asia_df['gross_inc_percap'].min()) /
                             (south_asia_df['gross_inc_percap'].max() - south_asia_df['gross_inc_percap'].min()))
```

Explanation: As life expectancy and GNI per capita are in different scales, they were normalized to a minmax normal and then the composite development score was constructed.

```python
south_asia_df['composite_score'] = (0.3 * south_asia_df['life_expectancy_index'] +0.3 * south_asia_df['gni_index'])   #creating and calculating Composite Development Score
south_asia_df[['country','composite_score','life_expectancy_index','gni_index']]




```

##### ii.Rank South Asian countries based on Composite Score.

```python
latest_year_df = (south_asia_df.sort_values('year').groupby('country').tail(1))
latest_year_df
```

As the dataset includes multiple years per country, using the full data caused the same country to appear repeatedly in the top five rankings. Therefore, only the most recent year of data for each country was selected to allow a valid cross-country comparison.

```python
ranked_df = (latest_year_df.sort_values('composite_score', ascending=False).reset_index(drop=True))
ranked_df['rank'] = ranked_df['composite_score'].rank(ascending=False,method='dense')

ranked_df[['rank', 'country', 'year', 'composite_score']]


```
